##**Step 1- Import Essential Library**

```python
import cv2
from google.colab.patches import cv2_imshow
```

##**Step 2- Load the Image**

```python
image = cv2.imread('shreeram.jpg')
```

##**Step 3- Convert image to grayscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

##**Step 4- Invert the Grayscale Image**

```python
inverted_gray_image = 255 - gray_image
```

##**Step 5- Apply Gaussian Blur to the Inverted Image**

```python
blurred_image = cv2.GaussianBlur(inverted_gray_image,(21, 21),0)
```

##**Step 6- Blend the Grayscale image with the blurred image using a color dodge blend mode**

```python
pencil_sketch = cv2.divide(gray_image,255 - blurred_image, scale=256)
```

##**Step 7- Show the Original Image and the pencil sketch one by one**

```python
cv2_imshow(image)
cv2_imshow(gray_image)
cv2_imshow(inverted_gray_image)
cv2_imshow(blurred_image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
